```python
import pandas as pd
from matplotlib import pyplot as plt
import os

import warnings
warnings.filterwarnings('ignore')
```

# Make dataset

```python
columns = ['fpkm_unstranded', 'fpkm_uq_unstranded']
genes = ['BRCA1', 'BRCA2']
genes_dict = {g: pd.DataFrame(columns=['id', *columns]) for g in genes}

path = './data'
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith('.tsv'):
            file_path = os.path.join(root, file)
            id = file.split('.')[0]
            try:
                df = pd.read_csv(file_path, sep='\t', comment='#')
                df = df[~df['gene_id'].isin(['N_unmapped', 'N_multimapping', 'N_noFeature', 'N_ambiguous'])]
                for g in genes:
                    b1 = df[df['gene_name']==g]
                    d = {'id':id}
                    for c in columns:
                        d[c]=b1[c]
                    
                    genes_dict[g] = pd.concat([genes_dict[g], pd.DataFrame(d)], ignore_index=True)

                
            except Exception as e:
                print(f"Errore nella lettura del file {file_path}: {e}")

genes_dict
```

```python
os.makedirs('./data/dataset', exist_ok=True)
for k, v in genes_dict.items():
    v.to_csv(f'./data/dataset/{k}.csv', index=False)
```

# Data visualization

```python
gene_name = 'BRCA1'
df = pd.read_csv(f'./data/dataset/{gene_name}.csv')
print(df)
```

```python

plt.figure(figsize=(10, 6))
plt.hist(df['fpkm_unstranded'], bins=20, alpha=0.5, label='fpkm_unstranded', color='blue')
plt.hist(df['fpkm_uq_unstranded'], bins=20, alpha=0.5, label='fpkm_uq_unstranded', color='orange')
plt.title(f'{gene_name}')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
min_val = df['fpkm_unstranded'].min()-1
max_val = df['fpkm_unstranded'].max()+1
step = 0.50
intervals = [round(min_val + i * step, 3) for i in range(int((max_val - min_val) / step) + 1)]
```

```python
df_cat = df.copy()
df_cat['fpkm_unstranded'] = pd.cut(df_cat['fpkm_unstranded'].values, intervals)
df_cat['fpkm_unstranded'].value_counts().sort_index().plot.bar()
```

```python
min_val = df['fpkm_uq_unstranded'].min()-1
max_val = df['fpkm_uq_unstranded'].max()+1
step = 0.50
intervals = [round(min_val + i * step, 3) for i in range(int((max_val - min_val) / step) + 1)]
```

```python
df_cat['fpkm_uq_unstranded'] = pd.cut(df_cat['fpkm_uq_unstranded'].values, intervals)
df_cat['fpkm_uq_unstranded'].value_counts().sort_index().plot.bar()
```
